```python
df=df.drop("state",axis=1)
df=df.drop("field38",axis=1)
df=df.drop("x",axis=1)
df=df.drop("schedule_id",axis=1)
df=df.drop("v126",axis=1)
```

```python
ds=df[["house_no","house_hold_no","district","rural","stratum_code","psu_id","m_serial_no","deceased_sex","m_expall_status","fid","year"]]
```

```python
ds["m_serial_no"].fillna(ds["m_serial_no"].mean(skipna=True),inplace=True)
ds["m_expall_status"].fillna(ds["m_expall_status"].mean(skipna=True),inplace=True)
ds["fid"].fillna(ds["fid"].idxmax(),inplace=True)
```

```python
ds=ds.astype({"m_serial_no":'int',"fid":'int'})
```

```python
cor2=ds.corr()
cor2.style.background_gradient(cmap='spring')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=ds.iloc[:,0:10]
y=ds["year"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model2=RandomForestClassifier()
model2.fit(x_train,y_train)
```

```python
pred1=model.predict(x_test)
print(pred1)
```

```python
pred2=model2.predict(x_test)
print(pred2)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc1=accuracy_score(y_test,pred1)
print("Accuracy:",acc1*100)
```

```python
acc2=accuracy_score(y_test,pred2)
print("Accuracy:",acc2*100)
```
